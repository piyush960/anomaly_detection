```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.ensemble import IsolationForest

%matplotlib inline
```

```python
features = ['amount', 'oldbalanceOrg', 'newbalanceOrig', 'oldbalanceDest', 'newbalanceDest']
df_final = df[features]
df_final['changeOrig'] = df['newbalanceOrig'] - df['oldbalanceOrg'] 
df_final['changeDest'] = df['newbalanceDest'] - df['oldbalanceDest']
df_final
```

```python
df_final['hour'] = df['step'] % 24
df_final
```

```python
type_one_hot = pd.get_dummies(df['type'])
df_train = pd.concat([df_final, type_one_hot], axis=1)
df_train.head()
```

```python
plt.scatter(df_train['amount'], df_train['newbalanceOrig'])
```

```python
plt.scatter(df_train['amount'], df_train['hour'])
```

```python
model = IsolationForest(n_estimators=100, contamination=0.0012, random_state=0)
model.fit(df_train)
```

```python
score = model.decision_function(df_train)
```
