### Feature Engineering

```python
features = df
features['transactionHour'] = features['step'] % 24

features['transactionBetween'] = df['nameDest'].apply(lambda x: 'Customer2Customer' if x[0] == "C" else 'Customer2Merchant')

features['transactionPeriod'] = df['transactionHour'].apply(lambda x: 'Peak' if x in [2,3,4,5,6] else
                                                                ('Mid' if x in [22,23,0,1,7,8] else 'Safe'))

features['origChange'] = features['newBalanceOrig'] - features['oldBalanceOrig']
features['destChange'] = features['newBalanceDest'] - features['oldBalanceDest']

features.head()
```

```python
features.drop(columns=['step', 'nameDest', 'nameOrig'], inplace = True)
features['transactionPeriod'].replace(['Safe', 'Mid', 'Peak'], [1, 2, 3], inplace=True)
features['transactionBetween'].replace(['Customer2Merchant', 'Customer2Customer'], [1, 2], inplace=True)
features['type'].replace(['CASH_IN','CASH_OUT','DEBIT','PAYMENT','TRANSFER'], ['1','2','3','4','5',], inplace=True)
features.head()
plt.figure(figsize =(10,10))
g=sns.heatmap(features.corr(),annot=True)
```

### Modelling

```python
x = features.drop(columns=['isFraud', 'isFlaggedFraud'])
y = features['isFraud']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state = 42, stratify=y)
```

```python
x_train.head()
```

```python
iso_forest = IsolationForest(n_estimators=100, random_state=0, verbose=0, contamination=0.00129)
```

```python
columns = x_train.columns.drop(['transactionPeriod', 'transactionBetween'])
iso_forest.fit(x_train[columns])
```

```python
score = iso_forest.decision_function(x_train[columns])
```

```python
threshold = 0.015
plt.hist(score, bins=35)
plt.axvline(threshold, color='r')
plt.show()
```

```python
x_train['score'] = pd.Series(score)
```

```python
merged_df = x_train.join(df['isFraud'])
merged_df.head()
```

```python
filt = merged_df['score'] < threshold
filt1 = merged_df['isFraud']==1
plt.scatter(x_train.loc[filt1, 'oldBalanceDest'], x_train.loc[filt1, 'amount'], edgecolors='black', alpha=0.5)
plt.scatter(x_train.loc[filt, 'oldBalanceDest'], x_train.loc[filt, 'amount'], edgecolors='black', alpha=0.8)
```
